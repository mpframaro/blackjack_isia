# 1. Resumo e Regras

O Blackjack é um dos jogos de sorte/azar mais conhecidos em todo o mundo. Merece esta fama, não só pelas suas simples regras, mas também por ser dos poucos jogos nos casinos que, sobre certas condições, permite ao jogador ter vantagem sobre a banca.<br>
Vamos rapidamente rever as regras implementadas nesta versão do jogo (simplificada para questões de optimização).

O jogo começa com o dealer tendo uma carta virada para cima e uma carta virada para baixo, enquanto que o jogador tem duas cartas viradas para cima. Todas as cartas são retiradas de um baralho infinito com reposição.<br>
O jogador tem a soma das cartas que possui, podendo decidir pedir mais cartas (1, hit) ou ficar com as suas cartas atuais (0, stick), até ultrapassar uma soma de 21 pontos (estourar, perda imediata).<br>
Depois de o jogador escolher stick, o dealer revela a sua carta virada para baixo e retira mais cartas para si até atingir uma soma igual ou superior a 17, mas igual ou inferior a 21 (se ultrapassar, o dealer estoura e ganha o jogador).<br>
Se ninguém estourar, ganha quem tiver uma soma mais perto de 21! Se houver empate, ninguém ganha.<br>
Esta versão do jogo tem algumas limitações, uma vez que não são implementadas ações como "separar", "dobrar", "seguros" e "desistências". A interface gráfica deste ambiente também não mostra a totalidade das cartas do dealer, estas apenas podem ser vistas através da função print.

Nota - as cartas numéricas têm um valor igual ao seu número (2-10), as cartas de figura (valete, dama, rei) valem 10 pontos, os ases podem valer 11 ou 1 pontos (será sempre contado como 11 a não ser que faça o jogador estourar).

# 2. Ambientes

Vamos agora ver o ambiente, aprender como funciona, e aplicar as mudanças que acharmos necessárias para melhor treinar os nossos agentes!

É importante notar que, para poder treinar os nossos agentes teremos de utilizar um wrapper que torne a nossa observação de um tuplo para uma box (um array), onde nos primeiros 32 valores (0-31) temos assinalado com 1 a soma das cartas do jogador, nos seguintes 11 valores (32-42, assumido como 0-11) temos assinalado com 1 a carta virada para cima da banca, e nos últimos 2 valores (43-44, assumido como 0-1) temos assinalado com 1 a presença de ás utilizável.<br>
Utilizaremos o FlattenObservation.

### 

### &emsp;&emsp;2.1 Ambiente Original

O ambiente que vamos utilizar ("Blackjack-v1") encontra-se disponível no github do Gymnasium (https://github.com/Farama-Foundation/Gymnasium).<br>
Iremos, ainda assim, inclui-lo no nosso trabalho (ficheiro blackjack.py), uma vez que foi utilizado como referência diversas vezes ao longo do desenvolvimento deste projeto, e fazemos uso direto dele, utilizando-o através de import.

Podemos ver que as ações tomam o valor 0 (stick) ou 1 (hit).<br>
As observações são tuplos com 3 valores. O primeiro valor representa a soma das cartas do jogador (varia entre 0 e 31), o segundo valor representa a carta virada para cima da banca (varia entre 0 e 10), e o terceiro valor representa a presença de um ás utilizável (toma o valor 0 para não e 1 para sim).<br>
Podemos concluir que há um total de 32 * 11 * 2 = 704 combinações de valores (estados ou observações) possíveis!

### &emsp;&emsp;2.2 Ambiente Modificado

Iremos agora modificar o ambiente de modo a ter um treino mais eficiente para os nossos agentes! Fizemos as seguintes mudanças:
- Penalizar agente por terminar o jogo com uma mão muito baixa (menos de 12 pontos);
- Beneficiar agente por ter uma mão numa posição muito vantajosa (entre 18 e 20 pontos);
- Penalizar agente por ter uma mão numa posição muito arriscada (mais de 20 pontos);
- Penalizar agente por ter demasiados "hits" numa só jogada (mais de 4 hits).

Todas estas mudanças foram realizadas na função "step", como mostrado abaixo, e as alterações foram guardadas (ficheiro customblackjack.py).<br>
Os valores fixos de 0.5 foram escolhidos em proporção com as restantes rewards, uma vez que estas mudanças não representam uma vitória (+1) nem uma derrota (-1), mas sim estratégias usadas para garantir um modo de jogo moderado e inteligente, isto é, sem muita ganância mas sem muita covardia.

# 3. Agentes, Treino e Hypertuning

Já vimos as regras do jogo e já modificamos o ambiente! Falta-nos apenas criar agentes e treiná-los. Antes de o fazer, vamos rapidamente dar uma olhada nas estatísticas deste jogo!
- Com uma estratégia básica, a banca tem uma vantagem entre 2%-3%, tendo o jogador uma taxa de vitória de cerca de 42%-45%.<br>
- Se o jogador se guiar por uma tabela de jogadas fixas, a vantagem da banca pode descer até 0.5%!<br>
- Por fim, um jogador experiente capaz de contar cartas muda todo este cenário e é capaz de ter vantagem sobre a casa entre os 0.5%-1.5%.

Ora, contar cartas neste cenário torna-se impossível pois estamos perante um baralho infinito com reposição. O nosso objetivo torna-se, portanto, treinar o nosso agente de modo a que ele crie uma tabela que atribui uma ação a cada combinação de soma de cartas existente (um eixo será a soma das cartas do jogador, enquanto que outro eixo será a carta virada para cima da banca; será criada uma tabela para quando existe ás usável e outra para quando não existe às usável).<br>
Podemos, assim, esperar que o nosso agente consiga atingir uma taxa de 45% de vitória.

Para termos um meio de comparação justo, iremos treinar dois agentes com algoritmos distintos, nos dois ambientes existentes (original e modificado).<br>
Começamos com o algoritmo Deep Q-Network (DQN), uma vez que é muito eficiente em espaços de ação discretos e tem uma complexidade baixa.<br>
Adicionamos o algoritmo Proximal Policy Optimization (PPO) uma vez que é o algoritmo de policy-gradient "state-of-the-art", muito robusto a ambientes ruidosos como blackjack.

Assim, ficamos com os seguintes agentes:
- Agente DQN treinado no ambiente original;
- Agente DQN treinado no ambiente modificado;
- Agente PPO treinado no ambiente original;
- Agente PPO treinado no ambiente modificado.

É importante notar que o nosso ambiente não é muito complexo. Tendo um espaço de observações de tamaho 704 e apenas 2 ações possíveis, espera-se que o treino dos agentes chegue a uma convergência  relativamente rapidamente (cerca de 10,000 timesteps).

De todo o modo, estabelecemos a seguinte regra: treinar cada agente por 10,000 timesteps durante um loop que se repete 100 vezes (totalizando 1,000,000 timesteps), guardando o modelo no final de cada loop! Assim, torna-se possível comparar a performance de modelos treinados com poucos timesteps face a modelos treinados com muitos timesteps.

# 4. Análise e Comparação dos Agentes

Neste passo, iremos utilizar os logs do tensorboard que guardamos para ver como correu o treino dos nossos agentes, comparando-os de acordo com a média das reward em cada o ambiente em que foram e em cada modelo.

Antes de mais, temos de definir uma função que aplique uma certa smoothess aos valores, de modo a melhor visualizar a evolução dos valores. Depois temos de extrair todos os dados necessários. Só depois podemos começar a obter os plots que desejamos.

```python
#função para aplicar smoothness aos dados
def exponential_moving_average(values, alpha=0.1):
    smoothed = []
    for value in values:
        if smoothed:
            smoothed.append(alpha * value + (1 - alpha) * smoothed[-1])
        else:
            smoothed.append(value)
    return smoothed
```

```python
#dados para agente DQN no ambiente original
dqnorig_dados="models/logs/DQNorig_0"

event_acc = EventAccumulator(dqnorig_dados)
event_acc.Reload()
events = event_acc.Scalars("rollout/ep_rew_mean")

steps_dqnorig = [e.step for e in events]
values_dqnorig = [e.value for e in events]

alpha = 0.01
smoothed_values_dqnorig = exponential_moving_average(values_dqnorig, alpha)
```

```python
#dados para agente DQN no ambiente modificado
dqnmod_dados="models/logs/DQNmod_0"

event_acc = EventAccumulator(dqnmod_dados)
event_acc.Reload()
events = event_acc.Scalars("rollout/ep_rew_mean")

steps_dqnmod = [e.step for e in events]
values_dqnmod = [e.value for e in events]

alpha = 0.01
smoothed_values_dqnmod = exponential_moving_average(values_dqnmod, alpha)
```

```python
#dados para agente PPO no ambiente original
ppoorig_dados="models/logs/PPOorig_0"

event_acc = EventAccumulator(ppoorig_dados)
event_acc.Reload()
events = event_acc.Scalars("rollout/ep_rew_mean")

steps_ppoorig = [e.step for e in events]
values_ppoorig = [e.value for e in events]

alpha = 0.01
smoothed_values_ppoorig = exponential_moving_average(values_ppoorig, alpha)
```

```python
#dados para agente PPO no ambiente modificado
ppomod_dados="models/logs/PPOmod_0"

event_acc = EventAccumulator(ppomod_dados)
event_acc.Reload()
events = event_acc.Scalars("rollout/ep_rew_mean")

steps_ppomod = [e.step for e in events]
values_ppomod = [e.value for e in events]

alpha = 0.01
smoothed_values_ppomod = exponential_moving_average(values_ppomod, alpha)
```

### 

### &emsp;&emsp;4.1 Resultados de Treino por Modelo

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))  

#DQN
axes[0].plot(steps_dqnorig, values_dqnorig, label="DQNorig (real)", color='navy', linewidth=0.5, alpha=0.25)
axes[0].plot(steps_dqnorig, smoothed_values_dqnorig, label="DQNorig (smoothed)", color='navy', linewidth=0.5)
axes[0].plot(steps_dqnmod, values_dqnmod, label="DQNmod (real)", color='dodgerblue', linewidth=0.5, alpha=0.25)
axes[0].plot(steps_dqnmod, smoothed_values_dqnmod, label="DQNmod (smoothed)", color='dodgerblue', linewidth=0.5)
axes[0].set_xlabel("Step")
axes[0].set_ylabel("Value")
axes[0].set_title("DQN Model Mean Reward per Step")
axes[0].legend()

#PPO
axes[1].plot(steps_ppoorig, values_ppoorig, label="PPOorig (real)", color='darkred', linewidth=0.5, alpha=0.25)
axes[1].plot(steps_ppoorig, smoothed_values_ppoorig, label="PPOorig (smoothed)", color='darkred', linewidth=0.5)
axes[1].plot(steps_ppomod, values_ppomod, label="PPOmod (real)", color='red', linewidth=0.5, alpha=0.25)
axes[1].plot(steps_ppomod, smoothed_values_ppomod, label="PPOmod (smoothed)", color='red', linewidth=0.5)
axes[1].set_xlabel("Step")
axes[1].set_ylabel("Value")
axes[1].set_title("PPO Model Mean Reward per Step")
axes[1].legend()

#plot
plt.tight_layout()
plt.show()
```

Podemos ver que para os agentes que utilizam o modelo **DQN**:
- a mean reward per episode começa um pouco baixa, mas **rapidamente converge** e **estabiliza entre as 50,000 e 100,000 timesteps**;<br>
- no agente treinado no **ambiente original, o valor estabiliza perto de -0.1**;
- no agente treinado no **ambiente modificado, o valor estabiliza perto de 0.05**.

Para os agentes que utilizam o modelo **PPO**:
- a mean reward per episode demora **mais tempo a convergir** e **estabiliza entre as 100,000 e 200,000 timesteps**;<br>
- no agente treinado no **ambiente original, o valor estabiliza perto de -0.1**;
- no moagentedelo treinado no **ambiente modificado, o valor estabiliza perto de 0.1**.

Caso jogassemos num **cenário mais realista**, onde houvesse um número limitado e não muito elevado de baralhos **(2-10 baralhos)**, torna-se-ia possível aplicar **estratégias de contagem de cartas** que dão a vantagem ao jogador (taxa de vitória superior a 50%).<br>

# 6. Interface Gráfica

Neste ambiente, a segunda carta do dealer nunca é mostrada no render!<br>
De qualquer modo, imprimimos todos os valores do jogo a acompanhar a interface gráfica.

```python
def usable_ace(hand):  # Does this hand have a usable ace?
    return int(1 in hand and sum(hand) + 10 <= 21)


def sum_hand(hand):  # Return current hand total
    if usable_ace(hand):
        return sum(hand) + 10
    return sum(hand)
```

```python
episodes=10
```

```python
#para visualizar mudar o ambiente
envmod_render = CustomBlackjackEnv(render_mode="human")
envmod_render = FlattenObservation(envmod_render)

episodios_render = 10
vitorias_render = 0
derrotas_render = 0
empates_render = 0

for ep in range(episodes):
    
    obs, info = envmod_render.reset()
    
    done = False
    truncated = False
        
    while not done and not truncated:
        
        action, _states = model_PPOmod.predict(obs)
        
        valor_jogador = sum_hand(envmod_render.env.player)
        valor_banca = envmod_render.env.dealer[0]
        as_usavel = usable_ace(envmod_render.env.player)
        
        print(f"Jogador: {valor_jogador}; Banca: {valor_banca}; Às usável: {as_usavel}")
        print(f"Action = {action}\n")
        
        obs, rewards, done, truncated, info = envmod_render.step(action)

        time.sleep(1)
        #envmod_render.render()
       
    valor_jogador = sum_hand(envmod_render.env.player)
    valor_banca = sum_hand(envmod_render.env.dealer)
    as_usavel = usable_ace(envmod_render.env.player)
    print(f"Jogador: {valor_jogador}; Banca: {valor_banca}; Às usável: {as_usavel}")
    
    if rewards>0:
        print("\nVITORIA! Reward:", rewards, "\n\n----------------\n")
        vitorias_render+=1
    elif rewards<0:
        print("\nDERROTA Reward:", rewards, "\n\n----------------\n")
        derrotas_render+=1
    elif rewards==0:
        print("\nEMPATE (ou má jogada)! Reward:", rewards, "\n\n----------------\n")
        empates_render+=1
    
envmod_render.close()
```
